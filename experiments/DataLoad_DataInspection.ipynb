```python
import pandas as pd
```

```python
#Loading all datasets into memory
items = pd.read_csv('../data/raw/olist_order_items_dataset.csv', encoding='utf-8')
sellers = pd.read_csv('../data/raw/olist_sellers_dataset.csv', encoding='utf-8')
leads = pd.read_csv('../data/raw/olist_marketing_qualified_leads_dataset.csv', encoding='utf-8')
deals = pd.read_csv('../data/raw/olist_closed_deals_dataset.csv', encoding='utf-8')
products = pd.read_csv('../data/raw/olist_products_dataset.csv', encoding='utf-8')
```

# DATAFRAME: Leads

VARIABLES

- **mql_id** -> Marketing Qualified Leads for OLIST marketplace
- **first_contact_date** -> Date of first sales contact 
- **landing_page_id** -> IDs for different landing pages (pseudo ID)
- **origin** -> Lead origin where seller was coming


```python
leads.info()
```

```python
leads.head(5)
```

```python
 counts_lp = leads.landing_page_id.value_counts().sort_index(ascending=False).sort_values(ascending=False)
 counts_lp.head(10).sort_values(ascending=True).plot.barh(figsize=(10,5))
```

```python
 counts_origin = leads.origin.value_counts().sort_index(ascending=False)
 counts_origin.sort_values(ascending=True).plot.barh(figsize=(10,5))
```
